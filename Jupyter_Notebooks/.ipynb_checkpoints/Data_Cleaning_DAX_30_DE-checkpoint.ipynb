```python
import pandas as pd
import numpy as np
```

```python
# retrieving raw data for the CAC 40 index into a DataFrame
df_dax_30 = pd.read_csv("/Users/jsialell/Data_Thief_Project_Team_Wall-e/data/ARCHIV/DAX_30_DE.csv")
```

```python
# renaming the columns of the DataFrame
df_dax_30.columns = ['date', 'dax_30_de_index']
```

```python
# changing the type from objet to datetime for the values in the date column
df_dax_30['date'] = pd.to_datetime(df_dax_30['date'])
```

```python
# keeping data from January 1st 2019 only
df_dax_30 = (df_dax_30[df_dax_30['date'] >= '2019-01-01']
             .reset_index())
```

```python
# sorting the DataFrame chronologically from the latest to the earliest data
df_dax_30 = df_dax_30.sort_values('date', ascending=False)
```

```python
# Setting new index and deleting unnecessary colums
df_dax_30 = (df_dax_30
             .reset_index()
             .drop(columns='index'))
```
