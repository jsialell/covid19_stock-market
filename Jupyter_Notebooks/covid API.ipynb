```python
import requests
import pandas as pd
```

1. Gathering data for China's first confirmed cases from Coronavirus COVID19 API

```python
url = "https://api.covid19api.com/dayone/country/china/status/confirmed"

resp = requests.get(url)
resp.json()
```

2. Getting date of first case into date format YY-MM-DD

```python
date = pd.DataFrame(resp.json()).iloc[0]["Date"]
date = pd.to_datetime(date, format="%Y-%m-%d")
date = date.strftime("%Y-%m-%d")
date
```

3. Defining a function that given a list of countries, uses Coronavirus' API to return a dictionary with the country and recordade first case date 

```python
def first_case(countries):
    first_case_dict = {}
    base_url = "https://api.covid19api.com/dayone/country/"
    for i in range(len(countries)):
        url = base_url + str(countries[i]) + "/status/confirmed"
        resp = requests.get(url)
        date = pd.DataFrame(resp.json()).iloc[0]["Date"]
        date = pd.to_datetime(date, format="%Y-%m-%d")
        date = date.strftime("%Y-%m-%d")
        first_case_dict[countries[i]] = date
    return first_case_dict
```

```python
countries = ["china", "italy", "japan", "germany", "france", "spain", "united-states"]
```

```python
first_cases = first_case(countries)
first_cases
```

4. Converting the dictionary into a DataFrame

```python
first_cases = pd.DataFrame(first_cases, index=range(1))
```

```python
first_cases = first_cases.transpose().rename(columns={0:"date"})
first_cases
```
