```python
dates_dict={}
for i in dates:
    dates_dict[i] = "sum"
```

```python
#dates_dict
```

```python
covid_country = covid_cases.groupby("Country/Region").agg(dates_dict).reset_index()

covid_country
```

```python
# getting the total number of cases per country to confirm if the sum of each country is equal to the original DataFrame
covid_country.sum(axis=1)
```
