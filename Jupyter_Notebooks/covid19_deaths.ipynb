```python
dates_dict={}
for i in dates:
    dates_dict[i] = "sum"
```

```python
deaths_country = covid_deaths.groupby("Country/Region").agg(dates_dict).reset_index()

deaths_country.head()
```

```python

deaths_country.sum(axis=1)
```

```python
# defining the total number of cases which is the last value from the table since it is cumulative data

deaths_country["Total"] = deaths_country["4/7/20"]
```

3. Defining focus countries

```python
countries = ["China", "France", "Germany", "Italy", "Japan", "Spain", "US"]
```

```python
deaths_focus_countries = deaths_country[deaths_country["Country/Region"].isin(countries)].reset_index().drop(columns="index")
```
