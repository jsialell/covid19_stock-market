3. defining a function in order to calculate the daily cases per country

```python
# creating a function

def new_cases_day(df, columns):
    df_new = pd.DataFrame(columns=columns)
    df_new[columns[0]] = df[columns[0]]
    for i in range(len(columns_dates)-1):
        df_new[columns_dates[i+1]] = df[columns_dates[i+1]] - df[columns_dates[i]]
    return df_new
```

```python
covid_new_cases = new_cases_day(covid_cases, columns_dates)
```

```python
# replacing the new values in the original data frame
for column in columns_dates:
    covid_cases[column] = covid_new_cases[column]
```

6. Creating a function in order to transpose and change date type to date format

```python
def transpose_dates(df):
    df = df.transpose()
    columns = list(df.iloc[0])
    df.columns = columns
    df = df.drop(index="country").reset_index().rename(columns={"index":"date"})
    df["date"] = pd.to_datetime(df["date"])
    return df
```

```python
covid_cases_cumulative = transpose_dates(covid_cases_cumulative)
```
